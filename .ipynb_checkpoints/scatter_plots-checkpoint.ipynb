```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.random.randn(100)
y = x + np.random.randn(100) + 10
print x
print y
```

```python
fig = plt.figure()
```

```python
plt.scatter(x, y, alpha=0.5, color='orchid')
fig.suptitle('Example Scatter Plot')
plt.grid(True)
```

```python
fig.savefig('./figs/example_scatter_plot.png')
#plt.show()
```
